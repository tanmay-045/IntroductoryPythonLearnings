### Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

## Matplotlib v/s Seaborn

| Feature                  | Matplotlib                                              | Seaborn                                              |
|--------------------------|---------------------------------------------------------|------------------------------------------------------|
| **Purpose**              | General-purpose plotting                                | Statistical data visualization                       |
| **Ease of Use**          | Requires more lines of code to achieve complex plots    | High-level interface with simpler syntax             |
| **Aesthetics**           | Basic and highly customizable                           | Attractive default themes and color palettes         |
| **Built-in Themes**      | Limited built-in themes                                 | Several built-in themes (e.g., darkgrid, whitegrid)  |
| **Built-in Datasets**    | No built-in datasets                                    | Provides built-in datasets for easy testing          |
| **Statistical Plots**    | Requires more effort and understanding to create        | Specialized functions for statistical plots          |
| **Complex Plots**        | Can create any type of plot but with more code          | Simplifies creation of complex plots like heatmaps   |
| **Customization**        | Highly customizable with detailed control               | Simplified customization with fewer parameters       |
| **Integration**          | Integrates well with other libraries like Pandas        | Built on top of Matplotlib, integrates with Pandas   |
| **Learning Curve**       | Steeper learning curve for beginners                    | Easier for beginners to create attractive plots      |
| **Annotations**          | Requires explicit coding for annotations                | Provides easier ways to annotate plots               |
| **Grid and Subplots**    | Detailed control over grid and subplot layouts          | FacetGrid and PairGrid for conditional plots         |
| **Functionality**        | Extremely versatile, used for a wide range of plots     | Designed for statistical and categorical data plots  |
| **Community and Support**| Large community with extensive documentation            | Growing community with good documentation            |


- **Matplotlib** is a powerful tool for creating a wide range of plots with detailed customization options. It's suitable for users who need fine-grained control over every aspect of their plots.
- **Seaborn** simplifies the creation of attractive and informative statistical graphics. It's ideal for users who need to quickly generate visually appealing plots with less code.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Seaborn comes with several built-in datasets that you can use to practice. One popular dataset is the iris dataset
iris = sns.load_dataset('iris')
```

```python
# head() function is used to display the first few rows of a DataFrame or Series. 
# By default, head() returns the first 5 rows,
# but you can specify the number of rows you want to see by passing a parameter to it.
iris.head()
```

```python
iris
```

```python
# scatter plot
# to visualize the relation between sepal_length & sepal_width using scatter plot 
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width, color='purple')
plt.xlabel("Length of Sepal")
plt.ylabel("Width of Sepal")
plt.show()
```

```python
# line plot
# to visualize the relation between petal length and petal width using line plot
sns.lineplot(x= iris.petal_length, y=iris.petal_width)
```

```python
# displot
# creates a distribution plot of the sepal_length column from the iris dataset.
sns.displot(iris['sepal_length'])
```

```python
# loading another dataset tips
tips= sns.load_dataset('tips')
```

```python
tips.head(6)
```

```python
# Histogram 
# What the Histogram Represents
# X-Axis (total_bill): Represents the total bill amounts in the dataset.
# Y-Axis (Count/Frequency): Represents the count or frequency of total bill amounts that fall into each bin.
# Bins: The histogram divides the range of total bill amounts into 30 bins. Each bin represents a range of bill amounts,
# and the height of the bin indicates how many bills fall into that range.


sns.histplot(data=tips, x='total_bill', bins=30)
```

```python
# Bar Plot
# Shows the relationship between a categorical variable and a continuous variable.

sns.barplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
# Box Plot
# Displays the distribution of data based on a five-number summary.

sns.boxplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
# Counting Categorical Values: 
# The primary purpose is to count the occurrences of each unique value in the categorical variable smoker.

tips['smoker'].value_counts()
```

```python
# replot
# The relplot function in Seaborn is used to plot data points and visualize relationships between variables.
sns.relplot( data = tips, x = tips.total_bill , y = tips.tip ,hue = 'smoker' )
```

```python
tips['size'].value_counts()
```

```python
sns.relplot( data = tips, x = tips.total_bill , y = tips.tip ,hue = 'size' )
```

```python
# But the above graph is little bit confusing 
# thats why we use 'style' in place of 'hue'
sns.relplot(data=tips, x = tips.total_bill, y=tips.tip, style='size')
```

```python
# We can use both at once
sns.relplot(data=tips, x = tips.total_bill, y=tips.tip, style='size',hue='smoker')
```

```python
# catplot
# catplot in Seaborn is designed to plot categorical data in various formats to show relationships between variables.

sns.catplot(x = 'day' , y = 'total_bill' , data = tips )
```

```python
sns.catplot(x = 'smoker' , y = 'tip' , data = tips )
```

```python
# joint plot
# A joint plot in Seaborn is a way to visualize the relationship between two numerical variables.
# It combines scatter plots for each variable and histograms or density plots along the axes.

sns.jointplot(x = tips.total_bill , y = tips.tip)
```

```python
# Pair Plot
# Shows pairwise relationships in a dataset.
sns.pairplot(tips)
```

A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. Heatmaps are useful for visualizing data trends, patterns, and correlations, especially in large datasets. Here’s a detailed explanation of how heatmaps work and how they are commonly used:

**Key Concepts**
**Matrix Representation:** Heatmaps typically visualize data in a matrix format where each cell represents a value. These values are represented using colors that vary in intensity or hue.

**Color Representation:** Colors in a heatmap encode the magnitude of values. For example, warmer colors (like red or yellow) often represent higher values, while cooler colors (like blue or green) represent lower values.

**Applications:** Heatmaps are commonly used in various fields such as:

**Data Analysis:** To explore correlations between variables.
**Biology and Medicine:** To visualize gene expression patterns or medical imaging data.
**Business and Finance:** To analyze stock market performance or customer behaviour.

```python
# Heatmap
# Displays a matrix where the individual values are represented by colors.

# tips.corr() computes the pairwise correlation of columns in the tips DataFrame.
corr = tips.corr()

# annot=True: Optional parameter that annotates each cell with the numeric value of the correlation coefficient.
# cmap='coolwarm': Specifies the color map (colormap) to be used for the heatmap. Here, 'coolwarm' is a diverging colormap that ranges from cool colors (e.g., blue) to warm colors (e.g., red), indicating negative and positive correlations, respectively.
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.show()
```

```python
# Facet grid
#  Multi-plot grid for plotting conditional relationships.

g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.histplot, "total_bill")
plt.show()
```

```python
# combining plots
sns.boxplot(data=tips, x='day', y='total_bill')
sns.stripplot(data=tips, x='day', y='total_bill', color='black', alpha=0.5)
plt.show()
```
