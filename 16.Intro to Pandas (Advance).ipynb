### Concatenate 

**The concat function in Pandas is used to concatenate or stack multiple DataFrame or Series objects along a particular axis (either rows or columns). This function is quite powerful and flexible, allowing you to combine data in various ways.**
<br>
syntax- `pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)`
<br>
- objs: A sequence or mapping of Series or DataFrame objects to concatenate.
- axis: The axis to concatenate along (0 for rows, 1 for columns). Default is 0.
- join: How to handle indexes on other axis. Options are 'outer' (default) or 'inner'.
- ignore_index: If True, do not use the index values along the concatenation axis. The resulting axis will be labeled 0, ..., n-1. Default is False.
- keys: If multiple levels are passed, should contain tuples. Construct hierarchical index using the passed keys as the outermost level.
- levels: Specific levels (unique values) to use for constructing a MultiIndex. Otherwise, they will be inferred from the keys.
- names: Names for the levels in the resulting hierarchical index.
- verify_integrity: Check whether the new concatenated axis contains duplicates. This can be very slow for large datasets. Default is False.
- sort: Sort non-concatenation axis if it is not already aligned when join='outer'. Default is False.
- copy: If False, do not copy data unnecessarily. Default is True.

```python
# concat vertically(Rowwise)

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5'],
    'D': ['D3', 'D4', 'D5']
})

df3 = pd.concat([df1, df2])
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# Concat Horizontally
df3 = pd.concat([df1, df2], axis =1)
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# Concatenating with Different Indexes and Ignore Index
df3 = pd.concat([df1, df2], ignore_index=True)
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
df4 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'E': ['E0', 'E1', 'E2']
})

df3 = pd.concat([df1, df4], join='inner')
print(df1,'\n\n')
print(df4,'\n\n')
print(df3)
```

```python
# observe this example below

# concat vertically(Rowwise)

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
})

df3 = pd.concat([df1, df2])
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

### Merge
**The merge function in Pandas is used to combine two DataFrames based on one or more common columns or indexes, similar to SQL joins. This function is very powerful for merging datasets in various ways, such as inner join, outer join, left join, and right join.**

syntax- `pandas.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)`
- left: A DataFrame.
- right: Another DataFrame.
- how: Type of merge to be performed:
  - 'left': Use only keys from left, similar to a SQL left outer join.
  - 'right': Use only keys from right, similar to a SQL right outer join.
  - 'outer': Use union of keys from both DataFrames, similar to a SQL full outer join.
- 'inner': Use intersection of keys from both DataFrames, similar to a SQL inner join. Default is 'inner'.
- on: Column or index level names to join on. Must be found in both DataFrames. If not provided, uses intersection of columns.
- left_on: Column or index level names to join on in the left DataFrame.
- right_on: Column or index level names to join on in the right DataFrame.
- left_index: If True, use the index (row labels) from the left DataFrame as the join key(s).
- right_index: If True, use the index (row labels) from the right DataFrame as the join key(s).
- sort: Sort the result DataFrame by the join keys in lexicographical order. Default is False.
- suffixes: A tuple of string suffixes to apply to overlapping columns. Default is ('_x', '_y').
- copy: If False, avoid copying data where possible. Default is True.
- indicator: If True, adds a column to output DataFrame called _merge with information on the source of each row.
- validate: If specified, checks if merge is of specified type.


```python
df1 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K4'],
    'C': ['C0', 'C1', 'C2', 'C4'],
    'D': ['D0', 'D1', 'D2', 'D4']
})

df3 = pd.merge(df1, df2, on='key')
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# left join
# similarly right can be performed
df3 = pd.merge(df1, df2, how='left', on='key')
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# outer join 
df3 = pd.merge(df1, df2, how='outer', on='key')
print(df3)
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# Merge on Multiple Columns

df3 = pd.DataFrame({
    'key1': ['K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0'],
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
})

df4 = pd.DataFrame({
    'key1': ['K0', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0'],
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
})

df5 = pd.merge(df3, df4, on=['key1', 'key2'])
print(df3,'\n\n')
print(df4,'\n\n')
print(df5)
```

```python
df6 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2'])

df7 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index=['K0', 'K2', 'K3'])

df8 = pd.merge(df6, df7, left_index=True, right_index=True, how='outer')
print(df6,'\n\n')
print(df7,'\n\n')
print(df8)
```

### join

**The join function in Pandas is used to combine two DataFrames based on their indexes or on a key column. It is a more straightforward way to merge datasets compared to the merge function, as it primarily focuses on joining based on the index or a specified column.**

syntax - `DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)`

- other: Another DataFrame to join with the caller.
- on: Column or index level names to join on. If not specified and the calling DataFrame is not on its index, other will join   on the index.
- how: How to handle the join operation:
  - 'left': Use calling frame’s index (or column if on is specified).
  - 'right': Use other’s index.
  - 'outer': Form the union of the calling frame’s index and other’s index.
- 'inner': Form the intersection of the calling frame’s index and other’s index.
- lsuffix: Suffix to use for overlapping column names in the left DataFrame. Default is an empty string.
- rsuffix: Suffix to use for overlapping column names in the right DataFrame. Default is an empty string.
- sort: Order result DataFrame lexicographically by the join keys. Default is False.

```python
# joining on indexes 

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2'])

df2 = pd.DataFrame({
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index=['K0', 'K2', 'K3'])

df3 = df1.join(df2)
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# left join 
df3 = df1.join(df2, how='left')
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# right join 
df3 = df1.join(df2, how='right')
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# outer join 
df3 = df1.join(df2, how='outer')
print(df1,'\n\n')
print(df2,'\n\n')
print(df3)
```

```python
# Joining on a Key Column

df4 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'E': ['E0', 'E1', 'E2'],
    'F': ['F0', 'F1', 'F2']
})

df3 = df1.join(df4.set_index('key'), on='A')
print(df1,'\n\n')
print(df4,'\n\n')
print(df3)
```

```python
# Handling Overlapping Column Names with Suffixes

df4 = pd.DataFrame({
    'A': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
}, index=['K0', 'K2', 'K3'])

df3 = df1.join(df4, lsuffix='_left', rsuffix='_right')
print(df1,'\n\n')
print(df4,'\n\n')
print(df3)
```

### apply

**The apply function in Pandas is a powerful tool used to apply a function along an axis (rows or columns) of a DataFrame or to each element of a Series. This function is very versatile and can be used for various purposes, such as data transformation, aggregation, and more.**

syntax- `DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), **kwds)
Series.apply(func, convert_dtype=True, args=(), **kwds)`

- func: The function to apply to each column/row.
- axis: The axis along which the function is applied:
  - 0 or 'index': apply the function to each column.
  - 1 or 'columns': apply the function to each row.
- raw: Determines if the function should receive the raw ndarray instead of the Series. Default is False.
- result_type: Used only when axis is 1 or 'columns'. It can be expand, reduce, or broadcast.
  - 'expand': List-like results will be turned into columns.
  - 'reduce': The result will be a Series.
  - 'broadcast': The result will be broadcast to the original shape of the DataFrame.
- args: Positional arguments to pass to the function.
- kwds: Additional keyword arguments to pass to the function.

```python
# Applying a Function to Each Column

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40],
    'C': [100, 200, 300, 400]
})

df1 = df.apply(lambda x: x + 1)
print(df,'\n\n')
print(df1)
```

```python
# Applying a Function to Each Row

df1 = df.apply(lambda x: x.sum(), axis=1)
print(df,'\n\n')
print(df1)
```

comparison between `apply` and `applymap` :

| Feature          | `apply`                                              | `applymap`                                           |
|------------------|------------------------------------------------------|------------------------------------------------------|
| **Function**     | Applies a function along an axis of the DataFrame    | Applies a function element-wise across the DataFrame |
| **Axis**         | Can be applied along rows (`axis=1`) or columns (`axis=0`) | Always applied element-wise                         |
| **Input**        | Series or DataFrame depending on the axis            | Always works on individual elements                  |
| **Output**       | Can return a Series, DataFrame, or scalar            | Always returns a DataFrame                           |
| **Usage**        | Useful for aggregations, transformations, or reductions on rows or columns | Useful for element-wise transformations              |
| **Example**      | `df.apply(np.sum, axis=0)`                           | `df.applymap(lambda x: x**2)`                        |
| **Flexibility**  | More flexible in terms of axis and output            | Less flexible, strictly element-wise operations      |
| **Common Use Cases** | Summing columns/rows, applying complex functions to rows/columns | Applying formatting, converting data types element-wise |

### Window functions

These functions are particularly useful for time series data, rolling calculations, and aggregations. Pandas provides several methods to apply window functions, such as rolling(), expanding(), and ewm() (Exponentially Weighted Windows).

```python
df= pd.DataFrame({'Col_A':[1,2,3,4,5,6],'Col_B':[4,8,12,16,20,24]})
```
