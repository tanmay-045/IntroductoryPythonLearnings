```python
m = [1,2,3,4,5,6]
n = [54,67,45,34,67,98]
```

```python
# this will create figure area of desired ' len * bre '
plt.figure(figsize=(4,2))   
# this will put graph points in scattered form having colour 'c=#223BDC', intensity(0.1-1.0) 'alpha=0.9', marker(through which fig you want to show points) ='x'
plt.scatter(m,n , c = '#223BDC',alpha=.9, marker='x') 
# to put label on x-axis
plt.xlabel("x-axis")
# to put label on y-axis
plt.ylabel("y-axis")
# to put title on graph
plt.title("X vs Y")
# to show grid on graph area
plt.grid()
```

```python
x1=np.linspace(1,10,100)
y1=np.sin(x1)
```

```python
x1
```

```python
y1
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x1,y1, c='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph of Sin(x)  (scattered points form)")
plt.show()

plt.figure(figsize=(6,3))
plt.plot(x1,y1, c='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph of Sin(x)   (continuos form)")
plt.show()
```

```python
plt.figure(figsize=(9,3))
plt.plot(x1,y1,'--r')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph of Sin(x)")
plt.show()
```

```python
# # A histogram in Matplotlib is a type of bar plot that represents the distribution of data
# # by showing the frequency of data points within specified ranges (bins). 
# # It is particularly useful for visualizing the distribution and spread of continuous data.

# Bins: The data range is divided into intervals called bins. Each bin represents the frequency of data points that fall within that interval.
# Frequency: The height of each bar in the histogram represents the number of data points within the corresponding bin.
# Density: Instead of frequency, the y-axis can represent the probability density, which normalizes the area under the histogram to 1.
```

```python
data_p=[1,3,6,3,1,9,6,3,5,7,3,5,7,4,2,1,7,4,9,8]
```

```python
plt.figure(figsize=(5,4))
plt.hist(data_p,edgecolor='black',color='#A1FA7A')
plt.xlabel("Number")
plt.ylabel("Frequency of number")
plt.title("Histogram")
plt.grid()
plt.show()
plt.savefig("Histogram.png")
```

```python
# 3D graphs

x3 = np.random.standard_normal(100)
y3 = np.random.standard_normal(100)
z3 = np.random.standard_normal(100)

# Create a figure and a 3D Axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the scatter plot
ax.scatter(x3, y3, z3)

# Set labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```

```python
# 3D line plot
t = np.linspace(0, 10, 100)
x4 = np.sin(t)
y4 = np.cos(t)
z4 = t

# Create a figure and a 3D Axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the line plot
ax.plot(x4, y4, z4)

# Set labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```
